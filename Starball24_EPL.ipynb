```python
import pandas as pd

df1 = pd.read_csv('https://drive.google.com/uc?export=download&id=1xS6KPcf90NAEDDKD7FrrU03CBnuNPgWt')
df2 = pd.read_csv('https://drive.google.com/uc?export=download&id=1OTWMDnjtOf_00-5TniNe7SpXoSu-Kjs7')
df = pd.concat([df1, df2], ignore_index=True)
df = df[['Date', 'Time','HomeTeam','AwayTeam','FTHG',	'FTAG',	'FTR',	'HTHG',	'HTAG',	'HTR', 'HS',	'AS',	'HST',	'AST',	'HF',	'AF',	'HC',	'AC',	'HY',	'AY',	'HR',	'AR' ]]
df.isnull().sum()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['AwayTeam_code'] = df['AwayTeam'].astype('category').cat.codes.astype("int64")
df['HomeTeam_code'] = df['HomeTeam'].astype('category').cat.codes.astype("int64")
df['FTR_code'] = df['FTR'].astype('category').cat.codes
df['HTR_code'] = df['HTR'].astype('category').cat.codes
df['Date'] = pd.to_datetime(df['Date'])
df['Day_code'] = df['Date'].dt.dayofweek.astype("int64")
df['Hour'] = df['Time'].str.replace(":.+", "", regex=True).astype("int")
```

```python
# Kamus untuk menyimpan pemetaan kode ke nama tim
team_code_map = dict(zip(df['FTR_code'], df['FTR']))
sorted_team_codes = sorted(team_code_map.items(), key=lambda item: item[0])
for code, team_name in sorted_team_codes:
    print(f"{code}: {team_name}")
```

```python
import sklearn

print(sklearn.__version__)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
train = df[df["Date"] < '2022-01-01']
test = df[df["Date"] > '2022-01-01']
predictors = ["HomeTeam_code","AwayTeam_code", "Hour", "Day_code"]
test
```

```python
rf.fit(train[predictors], train["FTR_code"])
```

```python
preds = rf.predict(test[predictors])
test[predictors]
```

```python
from sklearn.metrics import accuracy_score
accuracy =  accuracy_score(test["FTR_code"], preds)
print("Accuracy: ", accuracy)
```
